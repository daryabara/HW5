```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Создаём и выводим датасет из 10 строк, предсказание изменения цены
data = {
    'Площадь': [60, 70, 80, 80, 100, 65, 75, 85, 95, 105],
    'Количество комнат': [1, 2, 2, 3, 3, 1, 2, 2, 3, 3],
    'Цена': [6.1, 7.3, 8.1, 9.2, 10.0, 6.5, 6.8, 8.5, 9.6, 10.4]
}
#преобразую в таблицу
df = pd.DataFrame(data)

print("Выводим наши данные:")
print(df.to_markdown(index=False))
#записываю данные в csv файл
df.to_csv('price.csv', index=False)
print("\nДатасет сохранен в файлы: price.csv")

#Подготовка данных, выдлаю признаки (х) и переменную(у), также разделяем данные на обучающую и тестовую выборку 70/30
X = df[['Площадь', 'Количество комнат']]
y = df['Цена']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# инициализация моделей, выбрала 5: лнейная регрессия, дерево решений, случайный лес, ближайшие соседи, градиентный бустинг
models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree': DecisionTreeRegressor(max_depth=2, random_state=42),
    'Random Forest': RandomForestRegressor(n_estimators=5, random_state=42),
    'K-Neighbors': KNeighborsRegressor(n_neighbors=2),
    'Gradient Boosting': GradientBoostingRegressor(n_estimators=50, random_state=42)
}

# обучение и оценка
#список для хранения резуьтата
results = []
for name, model in models.items():
#обучение на тренировочных данных
    model.fit(X_train, y_train)
# предсказания для тест данных
    y_pred = model.predict(X_test)
#считаем метрики, название модели, средняя абсолютная ошибка, коэфф детерминации
    metrics = {
        'Model': name,
        'MAE': mean_absolute_error(y_test, y_pred),
        'R²': r2_score(y_test, y_pred)
    }
    results.append(metrics)

# сравнение моделей, сортируем модели по убыванию и выводим
results_df = pd.DataFrame(results).sort_values('R²', ascending=False)
print("\nСравнение моделей:")
print(results_df.to_markdown(index=False))

# график сравнения моделей
plt.figure(figsize=(18, 12))
#строим линию предсказаний
for model_name in models.keys():
    model = models[model_name]
    model.fit(X_train, y_train)
    plt.scatter(X_test['Площадь'], y_test, alpha=0.5, label='Реальные данные')
    plt.plot(X_test['Площадь'], model.predict(X_test), '--', linewidth=2, label=model_name)

plt.xlabel('Площадь')
plt.ylabel('Цена (млн руб)')
plt.title('Сравнение моделей')
plt.legend()
plt.grid(True)
plt.show()

# выбираем лучшей модели, где R самый большой
best_model_name = results_df.iloc[0]['Model']
best_model = models[best_model_name]
print(f"\nЛучшая модель: {best_model_name} (R² = {results_df.iloc[0]['R²']:.3f})")
```
